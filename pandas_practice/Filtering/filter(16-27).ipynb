```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'world_alcohol.csv')
df.head()
```

```python
#16
'''Write a Pandas program to filter those records 
where WHO region contains "Ea" substring from world alcohol consumption dataset.'''
df[df['WHO region'].str.contains('Ea',regex=False) ==True]
```

```python
#17
'''Write a Pandas program to filter those records where WHO region matches 
with multiple values (Africa, Eastern Mediterranean, Europe) '''
df[(df['WHO region']=='Africa') | (df['WHO region']=='Eastern Mediterranean') | (df['WHO region']=='Europe')]
```

```python
df[df['WHO region'].isin(['Africa', 'Eastern Mediterranean', 'Europe'])]
```

```python
#18
''' Write a Pandas program to filter those records which not appears in a given list. '''
df[~df['WHO region'].isin(['Africa', 'Eastern Mediterranean', 'Europe'])]
```

```python
#19
'''Write a Pandas program to filter all records
where the average consumption of beverages per person from .5 to 2.50'''
df[(df['Display Value']>0.5) & (df['Display Value']<2.5)]
```

```python
#20
'''Write a Pandas program to find average consumption of wine per person greater than 2'''
df[(df['Display Value']>2) & (df['Beverage Types']=='Wine')]
```

```python
#21
'''Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 '''
df.loc[::10]
```

```python
#22
'''Write a Pandas program to select consecutive columns and also 
select rows with Index label 0 to 9 with some columns'''
df.loc[:9]
```

```python
#23
'''Write a Pandas program to rename all and only some of the column names '''
df.rename(columns={'WHO region':'region','Beverage Types':'beverage'})
```

```python
#24
'''Write a Pandas program to find which years have all non-zero values
and which years have any non-zero values .'''
df[df['Year']!=0]
```

```python
#25
'''Write a Pandas program to filter all columns where all entries present, 
check which rows and columns has a NaN and finally drop rows with any NaNs'''
print('without Nan:')
df[~df.isnull().any(axis=1)]
```

```python
print('Has a Nan:')
df[df.isnull().any(axis=1)]
```

```python
#drop rows with any NaNs:
df = df.dropna()
```

```python
#26
'''Write a Pandas program to filter all records starting from the 'Year' column,
access every other column from world alcohol consumption dataset'''
df.loc[:,'Year'::2]
```
