```python
#4
'''Write a Pandas program to find and drop the missing values from World alcohol consumption dataset'''
print(df[df.isna().any(axis=1)])
print('count of missing values:')
print(df.isnull().sum())
df = df.dropna()
```

```python
#5
'''Write a Pandas program to remove the duplicates from 
'WHO region' column of World alcohol consumption dataset'''
df.drop_duplicates(subset='WHO region')
```

```python
#6
'''Write a Pandas program to find out the alcohol consumption
of a given year from the world alcohol consumption dataset'''
year = int(input('enter the year:'))
df[df.Year == year]
```

```python
#7
'''Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989'
from the world alcohol consumption dataset.'''
df[(df.Year==1987) | (df.Year==1989)]
```

```python
#8
'''Write a Pandas program to find out the alcohol consumption details 
by the 'Americas' in the year '1985'  from the world alcohol consumption dataset'''
df[(df['WHO region']=='Americas') & (df['Year']==1985)]
```

```python
#9
'''Write a Pandas program to find out the alcohol consumption details
in the year '1986' where WHO region is 'Western Pacific' and country is 'VietNam' from df'''
df[(df['Year']==1986) & (df['WHO region']=='Western Pacific') & df['Country']=='Viet Nam']
```

```python
#10
'''Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' 
where WHO region is 'Americas' from the world alcohol consumption dataset.'''
df[((df['Year']==1986) | (df['Year']==1989))  & (df['WHO region']=='Americas')]
```

```python
#11
''' Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' 
where WHO region is 'Americas' or 'Europe' from the world alcohol consumption dataset.'''
df[((df['Year']==1986) | (df['Year']==1989)) & ((df['WHO region']=='Americas') | (df['WHO region']=='Europe'))] 
```

```python
#12
'''Write a Pandas program to find out the 'WHO region, 'Country', 'Beverage Types' 
in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe' from df'''
df[((df['Year']==1986) | (df['Year']==1989)) & ((df['WHO region']=='Americas') | (df['WHO region']=='Europe'))][['WHO region','Country','Beverage Types']]
```

```python
#13
'''Write a Pandas program to find out the records where consumption of beverages per person average >=5 
and Beverage Types is Beer from world alcohol consumption dataset.'''
df[(df['Display Value']>=5) & (df['Beverage Types']=='Beer')]
```

```python
#14
'''Write a Pandas program to find out the records where consumption of beverages per person average >=4 
and Beverage Types is Beer, Wine, Spirits from world alcohol consumption dataset'''
df[(df['Display Value']>=5) & ((df['Beverage Types']=='Beer') | (df['Beverage Types']=='Wine') | (df['Beverage Types']=='Spirits'))]
```
