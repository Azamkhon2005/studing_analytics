```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']},
                  index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

```python
#16
'''Write a Pandas program to sort the DataFrame first by 'name' in descending order,
then by 'score' in ascending order'''
df.sort_values(by=['name','score'],ascending=[False,True])
```

```python
#17
'''Write a Pandas program to replace the 'qualify' column 
contains the values 'yes' and 'no' with True and False.'''
df['qualify'] = df['qualify'].replace({'yes':True,'no':False})
df
```

```python
#18
'''Write a Pandas program to change the name 'James' to 'Suresh' in name column of the DataFrame'''
df.loc[df['name']=='James','name'] = 'Suresh'
df
```

```python
#19
'''Write a Pandas program to delete the 'attempts' column from the DataFrame'''
df.drop('attempts',axis=1)
```

```python
#20
'''Write a Pandas program to insert a new column in existing DataFrame'''
color = ['Red','Blue','Orange','Red','White','White','Blue','Green','Green','Red']
df.insert(4,'color',color)
df
```

```python
#21
'''Write a Pandas program to iterate over rows in a DataFrame'''
exam_data = [{'name':'Anastasia', 'score':12.5}, 
             {'name':'Dima','score':9}, {'name':'Katherine','score':16.5}]
df1 = pd.DataFrame(columns=['name','score'])
for i in range(len(exam_data)):
    df1.loc[i] = exam_data[i]
df1
```

```python
#22
'''Write a Pandas program to get list from DataFrame column headers.'''
df.columns.to_list()
```

```python
#23
'''Write a Pandas program to rename columns of a given DataFrame'''
df1 = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6], 'col3': [7, 8, 9]})
df = df1.rename({'col1':'Column1','col2':'Column2','col3':'Column3'},axis=1)
```

```python
#24
'''Write a Pandas program to select rows from a given DataFrame'''
df = pd.DataFrame({'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]})
print('Rows for Column1 value == 4')
df[df['col1']==4]
```

```python
#25
''' Write a Pandas program to change the order of a DataFrame columns'''
df.sort_index(axis=1,ascending=False)
```

```python
#26
'''Write a Pandas program to add one row in an existing DataFrame'''
row = {'col1':10,'col2':11,'col3':12}
df = df.append(row, ignore_index = True)
df
```

```python
#27
'''Write a Pandas program to write a DataFrame to CSV file using tab separator'''
df.to_csv('new_file.csv',sep='\t')
new_df = pd.read_csv('new_file.csv')
new_df
```
