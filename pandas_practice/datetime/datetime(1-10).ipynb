```python
import pandas as pd
import numpy as np
data = pd.read_csv(r'ufo_sighting_data.csv')
data.rename({'Date_time' : 'date'},axis=1,inplace=True)
data['date'] = data['date'].str.replace('24:00', '0:00').astype('datetime64[ns]')
data.head()
```

```python
#1
'''Write a Pandas program to create the todays date'''
pd.to_datetime('today')
```

```python
#2
''' Write a Pandas program to calculate all the sighting days'''
data['date'].dt.day
```

```python
#3
'''Write a Pandas program to get the current date, 
oldest date and number of days between Current date and oldest date of Ufo dataset'''
now = pd.to_datetime('today')
old = data['date'].min()
now-old
```

```python
#4
import datetime
'''Write a Pandas program to get all the sighting days of the 
ufo which are less than or equal to 40 years (365*40 days)'''
data[now-data['date']<=datetime.timedelta(days=365*40)]
```

```python
#5
'''Write a Pandas program to get all the sighting days of the ufo between 1950-10-10 and 1960-10-10'''
data[('1950-10-10'<data['date'])& (data['date'] < '1960-10-10')]
```

```python
#6
'''Write a Pandas program to get all the sighting years of the (ufo) and create the year as column'''
data['year'] = data['date'].dt.year
data.head()
```

```python
#7
'''Write a Pandas program to create a plot to present the number of (UFO) reports per year.'''
import matplotlib.pyplot as plt
report = []
for i in range(data['date'].min().year,data['date'].max().year+1):
    report.append(data[data['date'].dt.year==i].shape[0])
plt.plot(list(range(data['date'].min().year,data['date'].max().year+1)),report)
```

```python
data['year'].value_counts().sort_index().plot()
```
