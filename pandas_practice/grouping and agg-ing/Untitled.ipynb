```python
data = pd.read_csv('aids.csv')
data.head()
```

```python
death = data.groupby('Year').agg({'Data.AIDS-Related Deaths.All Ages':'sum'},as_index=False)
```

```python
import matplotlib.pyplot as plt
death.plot()
plt.show()
```

```python
data.columns
```

```python
m_mean = data.groupby('Year').agg({'Data.New HIV Infections.Male Adults':'mean'},as_index=False)
f_mean = data.groupby('Year').agg({'Data.New HIV Infections.Female Adults':'mean'},as_index=False)
```

```python
plt.plot(m_mean,label = 'male')
plt.plot(f_mean,label = 'female')
plt.legend()
plt.show()
```
