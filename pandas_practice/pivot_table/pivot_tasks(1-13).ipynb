```python
import pandas as pd
import numpy as np
data = pd.read_excel('SaleData.xlsx')
data.head()
```

```python
#1
'''Write a Pandas program to create a Pivot table with multiple indexes from a given excel sheet'''
data.pivot_table(index=['Region','Item'])
```

```python
#2
''' Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise'''
data.pivot_table(index=['Region','Manager'],values='Sale_amt',aggfunc='sum')
```

```python
#3
'''Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise,
sales man wise'''
data.pivot_table(index=['Region','Manager','SalesMan'],values='Sale_amt',aggfunc='sum')
```

```python
#4
'''Write a Pandas program to create a Pivot table and find the item wise unit sold'''
data.pivot_table(index='Item',values='Units',aggfunc='sum')
```

```python
#5
'''Write a Pandas program to create a Pivot table and find the region wise total sale'''
data.pivot_table(index='Region',values='Units',aggfunc='sum')
```

```python
#6
''' Write a Pandas program to create a Pivot table and find the region wise, item wise unit sold'''
data.pivot_table(index=['Region','Item'],values='Units',aggfunc='sum')
```

```python
#7
'''Write a Pandas program to create a Pivot table and count the manager wise sale and mean value of sale amount'''
data.pivot_table(index='Manager',values='Sale_amt',aggfunc=['mean','count'])
```

```python
#8
'''Write a Pandas program to create a Pivot table and find manager wise, 
salesman wise total sale and also display the sum of all sale amount at the bottom. '''
data.pivot_table(index=['Manager','SalesMan'],values=['Sale_amt','Units'],aggfunc='sum',margins=True)
```

```python
#9
'''Write a Pandas program to create a Pivot table and find the total sale amount region wise,
manager wise, sales man wise where Manager = Douglas.'''
data.pivot_table(index=['Region','Manager','SalesMan'],values='Sale_amt',aggfunc='sum').query("Manager=='Douglas'")
```

```python
#10
''' Write a Pandas program to create a Pivot table and find the region wise
Television and Home Theater sold.'''
data[(data['Item']=='Television') | (data['Item']=='Home Theater')] \
.pivot_table(index='Region',values='Units')
```

```python
#11
'''Write a Pandas program to create a Pivot table and find the maximum sale value of the items'''
data.pivot_table(index='Item',values='Sale_amt',aggfunc='max')
```

```python
#12
'''Write a Pandas program to create a Pivot table and find the minimum sale value of the items'''
data.pivot_table(index='Item',values='Sale_amt',aggfunc='min')
```

```python
#13
'''Write a Pandas program to create a Pivot table and find the maximum and minimum sale value
of the items'''
data.pivot_table(index='Item',values='Sale_amt',aggfunc=['max','min'])
```
