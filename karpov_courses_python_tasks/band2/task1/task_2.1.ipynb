```python
'''7) –°–ª–µ–¥—É—é—â–∞—è –∑–∞–¥–∞—á–∞: –≤—ã—è—Å–Ω–∏—Ç—å, –∏–∑ –∫–∞–∫–æ–≥–æ —Ä–∞–π–æ–Ω–∞ –±—ã–ª–æ —Å–æ–≤–µ—Ä—à–µ–Ω–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ –∑–∞ –≤–µ—Å—å –ø–µ—Ä–∏–æ–¥. 
–î–ª—è –Ω–∞—á–∞–ª–∞, –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ (pickups), –±–µ–∑ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏.'''
df[df['pickups']==max(df['pickups'])]['borough']
```

```python
'''8) –ê —Ç–µ–ø–µ—Ä—å —Å–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Ä–∞–π–æ–Ω—É (borough) –∏ —É–∫–∞–∂–∏—Ç–µ, –∏–∑ –∫–∞–∫–æ–≥–æ –ø—É–Ω–∫—Ç–∞ –±—ã–ª–æ —Å–æ–≤–µ—Ä—à–µ–Ω–æ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø–æ–µ–∑–¥–æ–∫.'''
df.groupby('borough').agg({'pickups':'sum'}).idxmax()
```

```python
'''9) ‚≠êÔ∏è–ó–∞–¥–∞–Ω–∏–µ —Å–æ –∑–≤—ë–∑–¥–æ—á–∫–æ–π!‚≠êÔ∏è
–ï—â–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–ª–µ–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –≤ pandas! –í–æ–∑–º–æ–∂–Ω–æ, –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º —Å—Ç–µ–ø–µ –≤—ã –ø—Ä–æ—Å—Ç–æ –ø–æ—Å–º–æ—Ç—Ä–µ–ª–∏ –Ω–∞ –¥–∞–Ω–Ω—ã–µ –∏–ª–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è. –£—Å–∫–æ—Ä–∏—Ç—å –¥–∞–Ω–Ω—ã–π –ø—Ä–æ—Ü–µ—Å—Å –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –ø–æ–º–æ–≥—É—Ç –º–µ—Ç–æ–¥—ã idxmin() –∏ idxmax(), –∫–æ—Ç–æ—Ä—ã–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∏–ª–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.
–°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–∞–π–æ–Ω–∞ —Å –Ω–∞–∏–º–µ–Ω—å—à–∏–º —á–∏—Å–ª–æ–º –ø–æ–µ–∑–¥–æ–∫ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é min_pickups, –ø—Ä–∏–º–µ–Ω–∏–≤ –ø–æ–¥—Ö–æ–¥—è—â–∏–π –º–µ—Ç–æ–¥.'''
min_pickups = df.groupby('borough').agg({'pickups':'sum'}).idxmin()
```

```python
'''10) –ü—Ä–æ–¥–æ–ª–∂–∏–º –∏–∑—É—á–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ –≤ –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–Ω–∏.
–°–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –¥–≤—É–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º: —Ä–∞–π–æ–Ω—É –≥–æ—Ä–æ–¥–∞ –∏ —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–µ–Ω—å –≤—ã—Ö–æ–¥–Ω—ã–º (–∫–æ–ª–æ–Ω–∫–∏ borough –∏ hday).
–°—Ä–∞–≤–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫, –∏ –Ω–∞–∑–æ–≤–∏—Ç–µ —Ä–∞–π–æ–Ω—ã, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ –ø—Ä–∞–∑–¥–Ω–∏–∫–∞–º –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ—Å—Ç—É–ø–∞–µ—Ç –±–æ–ª—å—à–µ –∑–∞–∫–∞–∑–æ–≤, —á–µ–º –≤ –æ–±—ã—á–Ω—ã–µ –¥–Ω–∏. '''
mean_bor_day = df.groupby(by=['borough','hday'],as_index=False).agg({'pickups':'mean'})
mean_bor_day[mean_bor_day['hday']=='N'][mean_bor_day[mean_bor_day['hday']=='Y']['pickups'].values>mean_bor_day[mean_bor_day['hday']=='N']['pickups'].values]['borough']
```

```python
'''11) –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–∞–π–æ–Ω–∞ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ –ø–æ –º–µ—Å—è—Ü–∞–º. 
–û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –≤ pickups_by_mon_bor. 
–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∏—Ç–æ–≥–æ–≤—ã–π –¥–∞—Ç–∞—Å–µ—Ç –¥–æ–ª–∂–µ–Ω —Å–æ—Å—Ç–æ—è—Ç—å –∏–∑ 3-—Ö –∫–æ–ª–æ–Ω–æ–∫ ‚Äì pickup_month, borough, pickups.'''
pickups_by_mon_bor = df.groupby(['borough','pickup_month'],as_index=False).agg({'pickups':'sum'}).sort_values(by='pickups')
pickups_by_mon_bor
```
